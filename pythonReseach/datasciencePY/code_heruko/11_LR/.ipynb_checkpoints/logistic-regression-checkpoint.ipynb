```python
import pandas  as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("student_scores.csv")
df
```

```python
x = df['Hours']
y = df['Pass']
df.plot.scatter(x="Hours", y="Pass", c="Pass", colormap='bwr')
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def logistic_regression(all_x, all_y, m, b):
    for x,y_actual in zip(all_x, all_y):
        y_pred = sigmoid(m*x + b)
        error = y_pred - y_actual
        
        delta_m = -1*(error * x) * 0.05
        delta_b = -1 * (error) * 0.05
        
        m = m + delta_m
        b = b +  delta_b
        
    return m,b
```

```python
m = 0
b = 0
for i in range(0, 100):
    m,b = logistic_regression(x,y,m,b)
m,b
```

```python
 ### Helper function use when needed
def get_sigmoid(X, m, b):
    sig = []
    for x in X.values:
        y = sigmoid(m*x + b)
        sig.append(y)
        
    return sig
```

```python
plt.plot(x,y, 'o')
plt.xlabel("Hours")
plt.ylabel("Score")

sig = get_sigmoid(x, m, b)
plt.plot(x, sig, 'ro')
plt.show()
```

```python
def predict(x, m, b):
    y = sigmoid(m*x + b)
    if y > 0.5:
        return 1
    else:
        return 0
    
df["Predictions"] = df.apply(lambda row: predict(row["Hours"], m, b), axis=1)
df
```
