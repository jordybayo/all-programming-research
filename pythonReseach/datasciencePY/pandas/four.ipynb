# Manipulating the DataSet

```python
import pandas as pd
#read dataFrame and stor into a dataFrame
cars = pd.read_csv("GOOG.csv")
cars.head(6)
```

```python
# view Open fourth value
cars.Close[4]
# view only Close Stock
cars.iloc[:, 4] #iloc veut dire 'index location'
# same with ~ cars.Close[:]
```

```python
# see the first five reccord of Close column
cars.iloc[0:5, 4]
# same with ~ cars.Close.head(5)
```

```python
# view all rows and all columns
cars.iloc[:,:]
#same with ~ cars ~ car[:]
```

```python
# see records from a certain index
cars.iloc[6:, 4] # from 6th to all
```

```python
# see all the record of Close column 
cars.loc[:, "Close"] # loc veut dire 'location'
```

```python
# indexer en colonne les 7 premier ligne
cars.loc[:6, "Open":"Close"]
```

```python
# definir la valeur a la colonne Adj Close
cars['Adj Close'] = 1
cars
```

```python
# doubler les enregistrements de 'Adj Close' en utilisant les fonction lamda fxn
f = lambda x: x*2
cars['Adj Close'] = cars['Adj Close'].apply(f)
cars
```

```python
# trier une colonne en Ordre Ascendant
cars.sort_values(by='Volume')
```

```python
# trier une colonne en Ordre DSC
cars.sort_values(by='Volume', ascending=False)
```

```python
# filtrer des donner
cars['Open'] > 1000
```

```python
# filtrer les enreg avec plus de 1 500 000 en volume
filter1 = cars['Volume'] > 1500000
# appliqueer le  filtre au DataFrame
filtered_new = cars[filter1]
# consulter la nouvelle dataFrame
filtered_new
```

```python
#filtre des donnes avec plus de 900 Open et plus de 1500000 Volumes
filter2 = (cars["Open"]<800) & (cars["Volume"]>1500000)
#appliquer le filtre a la dataframe
filtered_review = cars[filter2]
#afficher les donnees
filtered_review
```
